```python
# 17. Build a dataset with every possible (UserID, ProductID) pair (cross join)

#transactions_products = pd.merge (transactions,products, on = 'ProductID')
import itertools

# Define function for cartesian join
def cartesian(df1, df2):
    rows = itertools.product(df1.iterrows(), df2.iterrows())

    df = pd.DataFrame(left.append(right) for (_, left), (_, right) in rows)
    return df.reset_index(drop=True)

```

```python
a = users.loc[:'UserID']
b = products.loc[:'ProductID']
c = cartesian(a,b) # Call cartesian function
```

```python
d = c[['UserID','ProductID']] # selective columns to a new dataframe
d
```

```python
# 18. Determine how much quantity of each product was purchased by each user

e = pd.merge(c,transactions,on=['UserID', 'ProductID'],how='outer')
f = e[['UserID','ProductID','Quantity']]
f = f.fillna(0) # convert NaN to 0.0
f
```
